```python
import numpy as np
import copy
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
suitImg = image.load_img("GAPDH.jpg", target_size=(255, 255))
```

```python
test_img = image.img_to_array(suitImg)
print(test_img.shape)
print(test_img)
```

```python
print(test_img[0])
```

```python
print(test_img[0].shape)
```

```python
print(test_img[0][0])
```

```python
print(test_img.shape)
```

```python
test2 = [[0 for col in range(255)] for row in range(255)]
for k in range(255):
    for j in range(255):
        test2[k][j] = test2[k][j] + (test_img[k][j][0]+test_img[k][j][1]+test_img[k][j][2])
```

```python
print(test2)
```

```python
num1 = test2[0][0]
num2 = test2[0][0]
for k in range(255):
    for j in range(255):
        if(num1 > test2[k][j]):
            num1 = test2[k][j]
print(num1)
```

```python
test2 = test2-num1
print(test2)
```

```python
test3 = [0 for row in range(255)]
print(test3)
```

```python
num = 0
for k in range(255):
    for j in range(255):
        num += test2[j][k]
    test3[k] = num
    num = 0
print(test3)
```

```python
num2=test3[0]
for k in range(255):
    if(num2 < test3[k]):
        num2 = test3[k]

print(num2)
```

```python
x = [a for a in range(1,256)]
y = [b for b in test3]
plt.plot(x,y)
plt.ylim(10, num2+10000)
plt.xlim(0, 256)
plt.show()
```

```python
TF = 0
for k in range(1, 255):
    if(TF == 1):
        test3[k] = 0
    if(test3[k-1]+1000 < test3[k]):
        TF = 1
```
